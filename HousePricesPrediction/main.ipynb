### Filling NA
Data suggest that 'BsmtQual', 'BsmtCond', 'BsmtFinType1' can be considered as 'NA' (No Basement) but in our records we checked there is no record of NA and instead we imply it was treated as NaN because all the values in these features where it is NaN are same.

```python
cols = ['BsmtQual', 'BsmtCond', 'BsmtFinType1']

train_hp[cols] = train_hp[cols].fillna('NA')

train_hp_records = train_hp.copy()
```

### Convert Categorical Data Into Nominal Or Ordinal Depends On The Condition
If the order lives only in your instinct and not in the data definition, dont force it on the model.

Example:

Neighborhood: A < B < C
But in reality:
A might beat B for small houses
B might beat C for old houses
C might only be premium with certain features
So the relationship is:
Contextual
Non-linear
Interaction-heavy
Trees can learn this  if you dont constrain them.

 One-Hot Encode: 
MSZoning, Neighborhood, Foundation, HouseStyle

 Ordinal Encode: 
ExterQual, BsmtQual, BsmtCond, BsmtFinType1, KitchenQual

```python
# First Convert Ordinal Ones
qual_map = {
    'NA': 0,
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

ordinal_cols = [
    'ExterQual',
    'BsmtQual',
    'BsmtCond',
    'KitchenQual'
]

for col in ordinal_cols:
    train_hp[col] = train_hp[col].map(qual_map)

bsmtfin_map = {
    'NA': 0,
    'Unf': 1,
    'LwQ': 2,
    'Rec': 3,
    'BLQ': 4,
    'ALQ': 5,
    'GLQ': 6
}

train_hp['BsmtFinType1'] = train_hp['BsmtFinType1'].map(bsmtfin_map)
```

```python
onehot_cols = [
    'MSZoning',
    'Neighborhood',
    'Foundation',
    'HouseStyle'
]

train_hp = pd.get_dummies(
    train_hp,
    columns=onehot_cols,
    drop_first=True
)

train_hp.head()
```

### First Model Implication (Random Forest Regression)

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Split features & target
X = train_hp.drop(columns=['SalePrice'])
y = train_hp['SalePrice']
# Create bins (10 is a good default)
y_bins = pd.qcut(y, q=5, labels=False)

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

train_rmse = []
val_rmse = []

rows = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y_bins), 1):

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # Random Forest
    model = RandomForestRegressor(
        n_estimators=300,
        max_depth=5,
        min_samples_leaf=30,
        random_state=1,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    # Predictions
    y_train_pred = model.predict(X_train)
    y_val_pred = model.predict(X_val)

    # RMSE
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_val = np.sqrt(mean_squared_error(y_val, y_val_pred))

    train_rmse.append(rmse_train)
    val_rmse.append(rmse_val)

    for idx, actual, pred in zip(
        X_val.index, y_val.values, y_val_pred
    ):
        rows.append([idx, actual, round(pred, 2), round(actual - pred, 2), fold])

    print(f"Fold {fold}")
    print(f"Train RMSE: {rmse_train:.2f}")
    print(f"Val   RMSE: {rmse_val:.2f}")
    print("-" * 30)

# Summary
print("Average Train RMSE:", np.mean(train_rmse))
print("Average Val   RMSE:", np.mean(val_rmse))
```

### Applying Log-transform 

```python
plt.figure(figsize=(12,5))

# Original
plt.subplot(1,2,1)
plt.hist(train_data['SalePrice'], bins=30)
plt.title("Original SalePrice")

# Log Transformed
plt.subplot(1,2,2)
plt.hist(np.log1p(train_data['SalePrice']), bins=30)
plt.title("Log(SalePrice)")

plt.show()
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Split features & target
X = train_hp.drop(columns=['SalePrice'])
y = np.log1p(train_hp['SalePrice'])
# Create bins (10 is a good default)
y_bins = pd.qcut(y, q=5, labels=False)

skf = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

train_rmse_log = []
val_rmse_log = []

train_rmse_actual = []
val_rmse_actual = []

all_actual = []
all_predicted = []

rows = []

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y_bins), 1):

    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]

    # Random Forest
    model = RandomForestRegressor(
        n_estimators=300,
        max_depth=None,
        min_samples_leaf=5,
        random_state=1,
        n_jobs=-1
    )

    model.fit(X_train, y_train)

    # Predictions Log
    y_train_pred_log = model.predict(X_train)
    y_val_pred_log = model.predict(X_val)


    # Convert actual values back to original scale
    y_train_actual = np.expm1(y_train)
    y_val_actual = np.expm1(y_val)

    # Convert predictions back
    y_train_pred = np.expm1(y_train_pred_log)
    y_val_pred = np.expm1(y_val_pred_log)

    # -------- LOG SCALE RMSE --------
    rmse_train_log = np.sqrt(mean_squared_error(y_train, y_train_pred_log))
    rmse_val_log = np.sqrt(mean_squared_error(y_val, y_val_pred_log))
    
    train_rmse_log.append(rmse_train_log)
    val_rmse_log.append(rmse_val_log)

    rmse_train_actual = np.sqrt(mean_squared_error(y_train_actual, y_train_pred))
    rmse_val_actual = np.sqrt(mean_squared_error(y_val_actual, y_val_pred))
    
    train_rmse_actual.append(rmse_train_actual)
    val_rmse_actual.append(rmse_val_actual)

    for idx, actual, pred in zip(
        X_val.index, y_val_actual.values, y_val_pred
    ):
        rows.append([idx, actual, round(pred, 2), round(actual - pred, 2), fold])

    all_actual.extend(y_val_actual.values)
    all_predicted.extend(y_val_pred)

    print(f"Fold {fold}")
    print(f"Train RMSE (Log): {rmse_train_log:.4f}")
    print(f"Val   RMSE (Log): {rmse_val_log:.4f}")
    print(f"Train RMSE (Actual): {rmse_train_actual:.2f}")
    print(f"Val   RMSE (Actual): {rmse_val_actual:.2f}")
    print("-" * 40)

# Summary
print("Average Train RMSE:", np.mean(train_rmse))
print("Average Val   RMSE:", np.mean(val_rmse))
```

### This feature has a little noise but also has a moderate upward direction, so it can be helpful in combination and we can try it in the model.

```python
# NULL LotFrontage groups
null_stats = (
    train_data.loc[train_data["LotFrontage"].isna()]
    .groupby(["Neighborhood", "LotShape"])
    .agg(
        null_count=("LotFrontage", "size"),
        LotArea_median_null=("LotArea", "median")
    )
)

# NON-NULL LotFrontage groups
non_null_stats = (
    train_data.loc[train_data["LotFrontage"].notna()]
    .groupby(["Neighborhood", "LotShape"])
    .agg(
        LotFrontage_median=("LotFrontage", "median"),
        LotFrontage_count=("LotFrontage", "count"),
        LotArea_median_non_null=("LotArea", "median")
    )
)

# Combine
result = (
    null_stats
    .join(non_null_stats)
    .sort_values("null_count", ascending=False)
)

result["LotArea_pct_diff"] = round(abs(
    (result["LotArea_median_null"] - result["LotArea_median_non_null"])
    / result["LotArea_median_non_null"]
) * 100, 2)

result
```

```python
neigh_stats = (
    train_data.loc[train_data["LotFrontage"].notna()]
    .groupby("Neighborhood")
    .agg(
        neigh_median_frontage=("LotFrontage", "median"),
        neigh_count=("LotFrontage", "count")
    )
)
result = result.join(neigh_stats, on="Neighborhood")
```

### Filling LotFrontage Data

```python
def classify_group(row):
    count = row["LotFrontage_count"]
    area_diff = row["LotArea_pct_diff"]
    neigh_count = row["neigh_count"]

    if pd.notna(count) and count >= 10 and area_diff <= 15:
        return "HIGH"
    elif pd.notna(count) and count >= 5 and area_diff <= 30:
        return "MEDIUM"
    elif pd.notna(count) and count >= 5:
        return "LOW_AREA_MISMATCH"
    elif pd.notna(neigh_count) and neigh_count >= 10:
        return "USE_NEIGHBORHOOD_MEDIAN"
    else:
        return "FALLBACK"

result["confidence"] = result.apply(classify_group, axis=1)
```

```python
import numpy as np

global_median = round(train_data["LotFrontage"].median(), 2)

def compute_imputation(row):
    conf = row["confidence"]
    median_front = row["LotFrontage_median"]
    null_area = row["LotArea_median_null"]
    non_null_area = row["LotArea_median_non_null"]
    neigh_median_frontage = row["neigh_median_frontage"]

    if conf == "HIGH":
        return round(median_front, 2)

    elif conf == "MEDIUM":
        if pd.notna(null_area) and pd.notna(non_null_area):
            # Why sqrt?
            # Because area is 2D and frontage is 1D.
            # This stabilizes scaling.
            scale = np.sqrt(null_area / non_null_area) 
            return round(median_front * scale, 2)
        else:
            return round(median_front, 2)

    elif conf == "LOW_AREA_MISMATCH":
        return round(median_front, 2)

    elif conf == "USE_NEIGHBORHOOD_MEDIAN":
        return round(neigh_median_frontage, 2)

    else:
        return global_median   # fallback will be handled later

result["imputed_preview"] = result.apply(compute_imputation, axis=1)
```

```python
def explain(row):
    return f"{row['confidence']} | count={row['LotFrontage_count']} | area_diff={row['LotArea_pct_diff']}%"

result["reason"] = result.apply(explain, axis=1)
```
